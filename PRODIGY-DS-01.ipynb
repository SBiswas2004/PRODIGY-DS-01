**Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.**



Sample Dataset :- https://data.worldbank.org/indicator/SP.POP.TOTL

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("Data/cleaned_data_Task1.csv")
df
```

```python
df = df.set_index("Country Name")
df.index
df
```

```python
print(df.index)
```

```python
col = df.columns.tolist()
print(col)
```

```python
for i in col:
        fig = plt.figure(figsize=(5,5))
        plt.hist(df[i],color='yellow',bins=10)
        plt.title(f"Population Distribution in {i}")
        plt.xlabel(i)
        plt.show()       
```

```python
df.sort_values(by='2022', ascending=False).head(10).plot(kind='bar', y='2022', title='Top 10 Populated Countries in 2022')
```

```python
years = df.columns 

total_values = df[years].sum()

bar_width = 0.5
bar_spacing = 0.5
bar_positions = np.arange(len(years)) * (bar_width + bar_spacing)

plt.figure(figsize=(15, 10))
plt.barh(bar_positions, total_values, height=bar_width, color='blue', align='center')
plt.yticks(bar_positions, years) 
plt.xlabel('Total Values',size=15)
plt.ylabel('Year',size=15)
plt.title('Total Values per Year',size=30)
plt.show()
```

```python
df_increase = df.diff(axis=1)

country = 'India'
country_data = df_increase.loc[country].dropna()

plt.figure(figsize=(10, 6))
plt.plot(country_data.index, country_data.values, marker='o')
plt.title(f'Year-over-Year Population Increase in {country}')
plt.xlabel('Year')
plt.ylabel('Population Increase')
plt.grid(True)
plt.show()
```

```python
total_population = df.sum(axis=1)
population_values = total_population.values

plt.figure(figsize=(12, 6))
plt.plot(total_population.index, population_values, marker='o')
plt.title('Total World Population Over Time')
plt.xlabel('Year')
plt.ylabel('Total Population')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
top_10_countries = df[['2022']].sort_values(by='2022', ascending=False).head(10)

plt.figure(figsize=(12, 8))
sns.barplot(data=top_10_countries, x='2022', y=top_10_countries.index)
plt.title('Top 10 Most Populated Countries in 2022')
plt.xlabel('Population')
plt.ylabel('Country')
plt.show()
```
